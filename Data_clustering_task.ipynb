```python
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
fig, ax = plt.subplots()
ax.scatter(X['sepal_length'], X['sepal_width'], c=kmeans.labels_, cmap='viridis')

ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_title('Iris Dataset: Sepal Length vs Sepal Width - Clusters')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
fig, ax = plt.subplots()
ax.scatter(X['petal_length'], X['petal_width'], c=kmeans.labels_, cmap='viridis')

ax.set_xlabel('petal_length (cm)')
ax.set_ylabel('petal_width (cm)')
ax.set_title('Iris Dataset: petal_length vs petal_width - Clusters')
plt.show()
```
